```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
url = 'https://doaj.org/api/search/articles/'
query = 'bibjson.abstract:ui AND bibjson.year:[2018 TO 2023]'
# query = 'bibjson.abstract:(("user\ interface" OR ui OR gui) AND (sketch OR image OR "mock\ up" OR mock-up OR prototype OR wireframe OR screenshot) AND ("automatic\ code\ generation" OR "automatic\ programming" OR "automatic\ code" OR "automatic\ software\ development" OR "to\ code")) AND (bibjson.year:[2018 TO 2023])'
```

```python
def obtain_link(dict_response_results):
    results = dict_response_results
    if results['bibjson']['link']:
        return  results['bibjson']['link'][0].get('url', '')
    else:
        return ''

def obtain_doi(dict_response_results):
    ids_list = dict_response_results['bibjson']['identifier']
    for id in ids_list:
        if id['type'] == 'doi':
            return id['id']

def obtain_authors(dict_response_results):
    result = dict_response_results
    author_list = []
    authors = result['bibjson'].get('author', '')
    
    if authors != '':
        for author in authors:
            author_list.append(author['name'])
        return author_list
    else:
        return authors

def iter_results(results_list_of_dicts):
    results = dict_response['results']
    
    for i, result in enumerate(results):
        print('Journal:', i)
        dict = {}
        
        dict['title'] = result['bibjson']['title']
        dict['abstract'] = result['bibjson']['abstract']
        dict['authors'] = obtain_authors(result)
        dict['year'] = result['bibjson']['year']
        dict['keywords'] = str(result['bibjson'].get('keywords', ''))
        dict['publication_type'] = ''
        dict['journal'] =  result['bibjson']['journal'].get('publisher', '') + \
        " - " + result['bibjson']['journal']['title']  + \
        " - " + result['bibjson']['journal'].get('volume', '') + \
        " - " + result['bibjson']['journal'].get('number', '')
        dict['article_url'] = obtain_link(result)
        dict['quotes_number'] = ''
        dict['doi'] = obtain_doi(result)
        dict['repository'] = 'DOAJ'
        dict['month'] = result['bibjson'].get('month', '')
        dict['query'] = dict_response['query'][18:]
        
        results_list_of_dicts.append(dict)
```

```python
# Consulto la primera página con la query definida.
response = requests.get(url + query)
dict_response = response.json()

page = dict_response['page']
total_results = dict_response['total']
print('Resultados totales', total_results)

results_list_of_dicts = []

# Validamos si el resultado tiene más de una página
if total_results <= 10:
    print('Página:', dict_response['page'])
# Si tiene una página, se genera el dataframe con los 10 resultados de esa página.Y los guarda en el results_list_of_dicts
    iter_results(results_list_of_dicts)
else:
# Si tiene más de una página, itero en cada una hasta la 100. Porque no puedo traer más de 1.000 resultados.    
    # Guardo los resultados de la primera página. 
    while page <= 100:
        print('Página:', dict_response['page'])
        # Itero los primeros 10 resultados y los guardo en el results_list_of_dicts.
        iter_results(results_list_of_dicts)
        # Consulto el link de la página siguiente.
        if dict_response['next']:
            next_page = dict_response['next']
        else:
            break
        
        # Llamo a la página siguiente.
        response = requests.get(next_page)
        dict_response = response.json()
       
        # Actualizo el número de página actual. 
        page = dict_response.get('page', 101)
        
print('Query finished.')
```
